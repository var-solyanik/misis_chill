```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def laplacian_blur_detection(image_path):
    """
    Функция для детекции размытости изображения с помощью оператора Лапласа.

    Args:
        image_path (str): Путь к входному изображению.

    Process:
        1. Загрузка изображения в оттенках серых.
        2. Применение оператора Лапласа к изображению.
        3. Вычисление статистики оператора Лапласа (среднего и стандартного отклонения).
        4. Анализ результатов для определения размытости.
        5. Визуальное представление исходного и обработанного изображений.
    """

    # Загрузка изображения в оттенках серых
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Преобразуем в оттенки серого
    
    # Применение оператора Лапласа
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    
    # Оцениваем статистику Лапласиана (среднее и стандартное отклонение)
    mean_laplacian = laplacian.mean()
    stddev_laplacian = laplacian.std()

    # Выводим результаты
    print(f"Среднее значение Лапласиана: {mean_laplacian}")
    print(f"Стандартное отклонение Лапласиана: {stddev_laplacian}")
    
    # Если изображение размыто, то Лапласиан будет иметь малые значения
    if mean_laplacian < -1:  # Пороговое значение зависит от изображения
        print("Изображение размыто")
    else:
        print("Изображение не размыто")

    # Отображаем оригинальное и обработанное изображение
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Оригинальное изображение")
    plt.imshow(image, cmap='gray')
    # plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Результат применения Лапласиана")
    plt.imshow(laplacian, cmap='gray')
    # plt.axis('off')

    plt.show()
```
